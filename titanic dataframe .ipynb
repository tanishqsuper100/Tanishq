```python
#now we are writting a funtion of age sex and children 
def child_male_female(passenger):
    age,sex=passenger
    if age<16:
        return "child"
    else:
        return sex
```

```python
#so we are aming new column 
titanic_df['person']= titanic_df[['Age',"Sex"]].apply(child_male_female,axis=1)
titanic_df.head(50)
```

```python
titanic_df # cabin column as many nan is there so lets dropn them 
```

```python
deck = titanic_df["Cabin"].dropna()
deck.head()
```

```python
levels = []
for x in deck:
    levels.append(x[0]) #we just need the first letter so [0]
    
levels
```

```python
#so lets viualize this by creating a data frame 
cabin_df = DataFrame(levels)
cabin_df.columns=['cabins'] #if you want to make a column you need to pass this []
sns.catplot(x='cabins',data=cabin_df,kind="count",palette="dark") 
```

```python
cabin_df.value_counts() # we cant drop T beacuse its not a column
```

```python
cabin_df=cabin_df[cabin_df.cabins != "T"] #that how you can drop a value in a cloumn without using a forloop 
sns.catplot(x='cabins',data=cabin_df,kind="count",palette="dark") 
```

who was alone and who was with family

```python
#if both SibSp	Parch	 are 0 the they alone 
titanic_df["alone"] = titanic_df['SibSp'] + titanic_df["Parch"]
#now lets make more than 0 as with family and 0 and no alone 
titanic_df['alone'].loc[titanic_df['alone']>0] = 'with family'#this one is nice to chnage a value withiout using for loop 
titanic_df['alone'].loc[titanic_df['alone']==0] = 'alone'
titanic_df
```

```python
sns.catplot(x='alone',data=titanic_df,kind='count')
```

```python
#which factor played different acts on survival 
#class was a factor 
sns.pointplot(x="Pclass",y='Survived',hue = 'person',data=titanic_df)
#yes it looks like pclass is cause of survival 
```
